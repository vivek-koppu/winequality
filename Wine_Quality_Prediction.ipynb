Wine Quality Prediction Using Support Vector Machine

The objective of a project is to find the quality of a wine

https://github.com/YBI-Foundation/Dataset/blob/main/WhiteWineQuality.csv

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/WhiteWineQuality.csv',sep=';')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.shape
```

```python
df['quality'].value_counts()
```

```python
df.groupby('quality').mean()
```

```python
y=df['quality']
```

```python
y.shape
```

```python
y
```

```python
x=df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
```

```python
x.shape
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s=StandardScaler()
```

```python
x=s.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,stratify=y,random_state=456)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

The project is done to check the quality of a wine. We have used the model called Support Vector Classifier. This model is used to train after training we have predicted for some values and we have printed the accuracy_score, classification_report, confusion_matrix.
